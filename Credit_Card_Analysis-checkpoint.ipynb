```python
df['age_group'] = pd.cut(
    df['customer_age'],
    bins=[20,30,40,50,60,80],
    labels=[
        '20-30',
        '31-40',
        '41-50',
        '51-60',
        '60+'
    ]
)
```

```python
df['utilization_pct'] = (
    df['avg_utilization_ratio'] * 100
)
```

```python
df['customer_segment'] = pd.qcut(
    df['total_trans_amt'],
    q=3,
    labels=[
        'Low Value',
        'Medium Value',
        'High Value'
    ]
)
```

```python
df['revenue_risk'] = np.where(
    df['attrition_flag'] ==
    'Attrited Customer',
    'High Risk',
    'Safe'
)
```

```python
churn_rate = (
    df['attrition_flag']
    .value_counts(normalize=True)
    *100
)
```
